```python
import requests
import json
from bs4 import BeautifulSoup
import os
import pandas
import urllib3
import random

l=[]
base_url="https://www.coursera.org/search?query=guided%20projects&page="
urlAfter = "&index=prod_all_products_term_optimization"

for page in range(1,33):
    
#     import requests
#     from bs4 import BeautifulSoup
#     import urllib3
#     import random
    user_agent_list = [
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.1.1 Safari/605.1.15',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0',
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
    ]
    user_agent = random.choice(user_agent_list)
    headers = {'User-Agent': user_agent}
    
    #check if the fake browsers and urls are working fine
    #i=int(page)
    #print("Request #%d\nUser-Agent Sent:%s\n\nHeaders Recevied by HTTPBin:"%(i,user_agent))
    url= base_url+str(page)+urlAfter
    print(url)
    res=requests.get(url,headers=headers)
    print(res.status_code)
    
    soup=BeautifulSoup(res.content, "html.parser")
    all=soup.find_all("li", {"class", "ais-InfiniteHits-item"})
    len(all)
    
    for i in range (0,int(len(all))):
        d={}
        try: d["Title"]=all[i].find("h2", {"class", "color-primary-text card-title headline-1-text"}).text
        except: d["Title"]= None
        try: d["Type"]=all[i].find("div", {"class", "_jen3vs _1d8rgfy3"}).text
        except: d["Type"]= None
        try: d["Rating score"]=all[i].find("span", {"class", "ratings-text"}).text
        except: d["Rating scor"]= None
        try: d["Rating Count"]=all[i].find("span", {"class", "ratings-count"}).span.text.replace(",","")
        except: d["Rating Count"]= None
        try: d["Enrollments"]=all[i].find("span", {"class", "enrollment-number"}).text
        except: d["Enrollments"]= None
        try: d["Level"]=all[i].find("span", {"class", "difficulty"}).text
        except: d["Level"]= None
        l.append(d)
```
